#####In December 2018, the team changed its name from the Delhi Daredevils to the Delhi Capitals
Sunrisers Hyderabad replaced the Deccan Chargers in 2012 and debuted in 2013 (But we consider them same in this kernel) .

```python
x=['Sunrisers Hyderabad', 'Mumbai Indians', 'Gujarat Lions',
    'Rising Pune Supergiant', 'Royal Challengers Bangalore',
    'Kolkata Knight Riders', 'Delhi Daredevils', 'Kings XI Punjab',
    'Chennai Super Kings', 'Rajasthan Royals', 'Deccan Chargers',
    'Kochi Tuskers Kerala', 'Pune Warriors', 'Rising Pune Supergiants', 'Delhi Capitals']

y = ['SRH','MI','GL','RPS','RCB','KKR','DC','KXIP','CSK','RR','SRH','KTK','PW','RPS','DC']

match_data.replace(x,y,inplace = True)
deliveries_data.replace(x,y,inplace = True)
```

```python
match_data.head(10)
```

```python
match_data['Season'] = pd.DatetimeIndex(match_data['date']).year
match_data.head()
```

```python
match_per_season=match_data.groupby(['Season'])['id'].count().reset_index().rename(columns={'id':'matches'})
match_per_season.style.background_gradient(cmap='spring')
```

### Importing matplotlib library


```python
import matplotlib.pyplot as plt
```

```python
x = match_per_season.Season
y = match_per_season.matches
plt.bar(x,y)
plt.show()
```

```python
x = match_per_season.Season
y = match_per_season.matches
plt.barh(x,y)
plt.show()
```

```python
x = match_per_season.Season
y = match_per_season.matches
plt.bar(x,y)
plt.xlabel("Season")
plt.ylabel("Matches")
plt.title("Number of matches played in different Seasons")
plt.show()
```

```python
x = match_per_season.Season
y = match_per_season.matches
plt.bar(x,y,color="green")
plt.xlabel("Season")
plt.ylabel("Matches")
plt.title("Number of matches played in different Seasons")
plt.show()
```

```python
x = match_per_season.Season
y = match_per_season.matches
plt.bar(x,y)
plt.xlabel("Season")
plt.ylabel("Matches")
plt.title("Number of matches played in different Seasons")
plt.grid(axis = 'y', color = 'red', linestyle = '--', linewidth = 0.5)
plt.show()
```
